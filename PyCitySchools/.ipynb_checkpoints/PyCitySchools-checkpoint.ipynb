```python
import os
import csv
import pandas as pd

```

```python
load_school_data = "Resources/schools_complete.csv"
load_students_data = "Resources/students_complete.csv"
```

```python
school_data = pd.read_csv(load_school_data)

stu_data = pd.read_csv(load_students_data)



school_set = pd.merge(stu_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
school_data.head()
```

```python
stu_data.head()
```

```python
school_set.head()
```

```python
school_set.describe()
```

```python
school_set.columns
```

```python
#calculate total schools
total_schools = len(school_set['school_name'].unique())
total_schools
```

```python
#calculate total students
total_students = school_set.count()['Student ID']
total_students
```

```python
#calculate total budget
total_budget = school_data.sum()['budget']
total_budget
```

```python
#calculate averge reading and math scores
reading_avg = school_set['reading_score'].mean()
math_avg = school_set['math_score'].mean()

print(reading_avg)
print(math_avg)
```

```python
#get passing math score
pass_math_score = school_set.loc[school_set['math_score']>=70]
total_pass_math_score = pass_math_score['student_name'].size
total_pass_math_score
```

```python
#get passing reading score
pass_read_score = school_set.loc[school_set['reading_score']>=70]
total_pass_read_score = pass_read_score['student_name'].size
total_pass_read_score
```

```python
#calculate percentage of math and reading scores individually 
passing_math_perc = total_pass_math_score / total_students * 100
passing_math_perc
```

```python
passing_read_perc = total_pass_read_score / total_students * 100
passing_read_perc
```

```python
#calculate overall passing of both subjects
overall_pass_score = school_set.loc[(school_set["math_score"]>=70) & (school_set["reading_score"]>=70)]

total_overall_score = overall_pass_score['student_name'].size

#calculate overall percentage of passingg scores
overall_perc = total_overall_score / total_students * 100

overall_perc
```

```python


new_df = {'Total Students':total_students,
          'Total Schools':total_schools,
          'Total Budget':total_budget,
          'Math Avg':math_avg,
          'Reading_Avg':reading_avg, 
          '% Passing_Math_Avg':passing_math_perc, 
          '% Passing_Read_Avg':passing_read_perc, 
          '% Overall_Passing_Percentage':overall_perc
}


new_df

```

```python
#create new dataframe with above data
#code inspired by classmate
dist_summary_df = pd.DataFrame([[total_schools, total_students, total_budget, math_avg, reading_avg, passing_math_perc, passing_read_perc, overall_perc]])
dist_summary_df
```

```python
dist_summary_df.columns =['Total School', 'Total Students','Total Budget', 'Math Average', 'Reading Average', '% Passing Math','% Passing Reading','Overall Percentage']
dist_summary_df.head()

```

```python
#format table for total studenta
dist_summary_df['Total Students'] = dist_summary_df['Total Students'].map("{:,}".format)
```

```python
#format table for total budget
dist_summary_df['Total Budget'] = dist_summary_df['Total Budget'].map("${:,.2f}".format)
```

# SCORES BY SCHOOL SPENDING


```python
bins = [0, 584.99, 629.99, 644.99, 680.99]
labels = ['<$585','$585-630', '$630-645', '$645-680']
```

```python
school_spending = school_set.merge(stu_budget.to_frame(), left_on='school_name', right_index=True, how='inner')
school_spending
```

```python
school_spending.rename(columns={0:'spending_per_school'}, inplace=True)
school_spending
```

```python
spend_df = pd.cut(school_spending['spending_per_school'], bins, labels=labels)

spend_df
```

```python
school_spending['bins'] = spend_df
school_spending.head()
```

```python
school_spending = school_spending.groupby(['bins']).mean()['spending_per_school']

school_spending
```

```python
school_sum_df['Spending Ranges (Per Student)'] = pd.cut(school_sum_df['Per Student Budget'], bins, labels=labels)

school_sum_df

```

```python
#create organized columns by averages

budget_chart_math = school_sum_df.groupby(["Spending Ranges (Per Student)"]).mean()["Average Math Score"]
budget_chart_reading = school_sum_df.groupby(["Spending Ranges (Per Student)"]).mean()["Average Reading Score"]
budget_chart_perc_math = school_sum_df.groupby(["Spending Ranges (Per Student)"]).mean()["% Passing Math"]
budget_chart_perc_reading = school_sum_df.groupby(["Spending Ranges (Per Student)"]).mean()["% Passing Reading"]
budget_chart_overall_pass = school_sum_df.groupby(["Spending Ranges (Per Student)"]).mean()["% Overall Passing"]





```

```python
#create budget chart dict

budget_chart_dict= {"Average Math Score":budget_chart_math,
                     "Average Reading Score":budget_chart_reading,
                     "% Passing Math":budget_chart_perc_math,
                     "% Passing Reading":budget_chart_perc_reading,
                     "% Overall Passing":budget_chart_overall_pass}
```

```python
budget_chart_df = pd.DataFrame(budget_chart_dict)
budget_chart_df

```

# SCORES BY SCHOOL SIZE

```python
#create bins
size_bins = [0, 1000, 2000, 5000]
size_labels = ["Small(<1000)" , "Medium(1000-2000)", "Large(2000-5000)"]
```

```python
stu_per_school
```

```python
school_size_df = school_data.merge(stu_per_school.to_frame(), left_on='school_name', right_index=True, how="left")

school_size_df
```

```python
school_sum_df["School Size"] = pd.cut(school_sum_df["Total Students"], bins=size_bins, labels=size_labels)


school_sum_df
```

```python
school_sum_df.head()
```

```python
#organize school size by scores


size_math_scores = school_sum_df.groupby(["School Size"]).mean()["Average Math Score"]

size_read_scores = school_sum_df.groupby(["School Size"]).mean()["Average Reading Score"]

size_math = school_sum_df.groupby(["School Size"]).mean()["% Passing Math"]

size_reading = school_sum_df.groupby(["School Size"]).mean()["% Passing Reading"]

size_passing = school_sum_df.groupby(["School Size"]).mean()["% Overall Passing"]

```

```python
# put size into a dataframe. 
school_size_df = pd.DataFrame({
          "Average Math Score" : size_math_scores,
          "Average Reading Score": size_read_scores,
          "% Passing Math": size_math,
          "% Passing Reading": size_reading,
          "% Overall Passing": size_passing})

school_size_df
```
